Baza danych: `ihrf`


```

import mysql.connector
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

DB_NAME = "ihrf"
CONFIG = {
    "host": "localhost",
    "user": "root",
    "password": "root_password",
    "database": DB_NAME,
}

conn = mysql.connector.connect(**CONFIG)
print("PoÅ‚Ä…czono z bazÄ…")
```

## Wczytanie danych z bazy

```

def read_sql(q):
    return pd.read_sql(q, conn)

employees = read_sql("SELECT * FROM employees")
hamsters = read_sql("SELECT * FROM hamsters")
disciplines = read_sql("SELECT * FROM disciplines")
competitions = read_sql("SELECT * FROM competitions")
doping_controls = read_sql("SELECT * FROM doping_controls")

for df in [employees, hamsters, competitions, doping_controls]:
    for c in df.columns:
        if "date" in c:
            df[c] = pd.to_datetime(df[c])
```

## Widok analityczny (JOIN)

```

comp = competitions.merge(disciplines, on="discipline_id")                    .merge(hamsters, on="hamster_id", suffixes=("", "_hamster"))

comp["year"] = comp["competition_date"].dt.year
comp.head()
```

## Najlepszy chomik w kaÅ¼dej kategorii

```

def higher_better(name):
    return "endurance" in name.lower()

comp["higher_better"] = comp["name"].apply(higher_better)

comp["z"] = comp.groupby("discipline_id")["winner_result"].transform(
    lambda s: (s - s.mean()) / (s.std() if s.std() > 0 else 1)
)

comp["score"] = np.where(comp["higher_better"], comp["z"], -comp["z"])

ranking = comp.groupby(["category", "hamster_id", "name_hamster"], as_index=False).agg(
    wins=("competition_id", "count"),
    avg_score=("score", "mean")
)

ranking["combined"] = ranking["wins"] + 0.5 * ranking["avg_score"]

ranking.sort_values(["category", "combined"], ascending=[True, False])        .groupby("category").head(1)
```

## Szczyt kariery chomikÃ³w

```

career = comp.groupby(["hamster_id", "name_hamster"], as_index=False).agg(
    wins=("competition_id", "count"),
    first=("competition_date", "min"),
    last=("competition_date", "max")
)

career = career[career["wins"] >= 2]
career["peak_days"] = (career["last"] - career["first"]).dt.days

career["peak_days"].describe()
```

## WpÅ‚yw majÄ…tku wÅ‚aÅ›ciciela na wyniki

```

rho, p = stats.spearmanr(comp["owner_networth"], comp["score"])
print("Spearman rho:", rho, "p-value:", p)
```

## PopularnoÅ›Ä‡ sportu (trend)

```

pop = comp.groupby("year").agg(
    events=("competition_id", "count"),
    hamsters=("hamster_id", "nunique"),
    avg_audience=("spectators_stadium", "mean")
)

pop.plot(subplots=True, figsize=(6,6), marker="o")
plt.tight_layout()
plt.show()
```

## Doping w czasie

```

doping_controls["year"] = doping_controls["control_date"].dt.year
doping = doping_controls.groupby("year").agg(
    tests=("control_id", "count"),
    positives=("is_positive", "sum")
)
doping["rate"] = doping["positives"] / doping["tests"]
doping
```
