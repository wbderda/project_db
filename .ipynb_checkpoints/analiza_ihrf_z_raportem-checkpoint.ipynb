# Analiza danych ‚Äì IHRF

Baza danych: `ihrf`


```python
import warnings
import mysql.connector
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

DB_NAME = "ihrf"
CONFIG = {
    "host": "localhost",
    "user": "root",
    "password": "root_password",
    "database": DB_NAME,
}

conn = mysql.connector.connect(**CONFIG)
print("Po≈ÇƒÖczono z bazƒÖ")
```

## Wczytanie danych z bazy

```python
warnings.filterwarnings(
    "ignore",
    message="pandas only supports SQLAlchemy connectable"
)
def read_sql(q):
    return pd.read_sql(q, conn)

employees = read_sql("SELECT * FROM employees")
hamsters = read_sql("SELECT * FROM hamsters")
disciplines = read_sql("SELECT * FROM disciplines")
competitions = read_sql("SELECT * FROM competitions")
doping_controls = read_sql("SELECT * FROM doping_controls")

for df in [employees, hamsters, competitions, doping_controls]:
    for c in df.columns:
        if "date" in c:
            df[c] = pd.to_datetime(df[c])
```

## Widok analityczny (JOIN)

```python

comp = competitions.merge(disciplines, on="discipline_id").merge(hamsters, on="hamster_id", suffixes=("", "_hamster"))

comp["year"] = comp["competition_date"].dt.year
comp
```

## Najlepszy chomik w ka≈ºdej kategorii

```python
# =========================================================
# Najlepszy chomik w ka≈ºdej dyscyplinie (discipline_id)
# ZA≈ÅO≈ªENIE: winner_result to CZAS, zawsze mniejszy = lepszy
# =========================================================

# 1) Normalizacja wynik√≥w w obrƒôbie dyscypliny (z-score)
comp["z"] = comp.groupby("discipline_id")["winner_result"].transform(
    lambda s: (s - s.mean()) / (s.std(ddof=0) if s.std(ddof=0) > 0 else 1.0)
)

# 2) Score: im wiƒôkszy, tym lepiej -> odwracamy znak, bo mniejszy czas jest lepszy
comp["score"] = -comp["z"]

# 3) Ranking chomik√≥w w ka≈ºdej dyscyplinie
rank_disc = (
    comp
    .groupby(
        ["discipline_id", "name", "category", "hamster_id", "name_hamster"],
        as_index=False
    )
    .agg(
        wins=("competition_id", "count"),
        avg_score=("score", "mean"),
        best_time=("winner_result", "min"),      # najlepszy (najmniejszy) czas
        avg_time=("winner_result", "mean"),
        first_win=("competition_date", "min"),
        last_win=("competition_date", "max")
    )
)

# 4) Metryka ≈ÇƒÖczna: liczba zwyciƒôstw + jako≈õƒá wynik√≥w (wins wa≈ºniejsze)
rank_disc["combined"] = rank_disc["wins"] + 0.5 * rank_disc["avg_score"]

# 5) Najlepszy chomik dla ka≈ºdej dyscypliny
best_by_discipline = (
    rank_disc
    .sort_values(["discipline_id", "combined"], ascending=[True, False])
    .groupby("discipline_id")
    .head(1)
    .sort_values("discipline_id")
)

# 6) Wynik ko≈Ñcowy (tabela do raportu)
best_by_discipline[
    [
        "discipline_id",
        "name",
        "category",
        "hamster_id",
        "name_hamster",
        "wins",
        "best_time",
        "avg_time",
        "avg_score",
        "first_win",
        "last_win",
    ]
]
```

## Szczyt kariery chomik√≥w

```python
# =========================================================
# Ile trwa "szczyt kariery" u chomik√≥w?
# Definicja: last_win - first_win (dla chomik√≥w z >=2 zwyciƒôstwami)
# =========================================================

career = (
    comp.groupby(["hamster_id", "name_hamster"], as_index=False)
        .agg(
            wins=("competition_id", "count"),
            first_win=("competition_date", "min"),
            last_win=("competition_date", "max")
        )
)

career2 = career[career["wins"] >= 2].copy()
career2["peak_days"] = (career2["last_win"] - career2["first_win"]).dt.days
career2["peak_months"] = career2["peak_days"] / 30.44

print("Liczba chomik√≥w z >=2 zwyciƒôstwami:", len(career2))
print("≈öredni szczyt kariery:", round(career2["peak_days"].mean(), 1), "dni (",
      round(career2["peak_months"].mean(), 2), "mies.)")
print("Mediana szczytu kariery:", round(career2["peak_days"].median(), 1), "dni (",
      round(career2["peak_months"].median(), 2), "mies.)")
print("25%‚Äì75%:", career2["peak_days"].quantile(0.25), "‚Äì", career2["peak_days"].quantile(0.75), "dni")

# rozk≈Çad
import matplotlib.pyplot as plt
plt.figure()
plt.hist(career2["peak_days"], bins=15)
plt.title("Rozk≈Çad d≈Çugo≈õci 'szczytu kariery' (dni)")
plt.xlabel("dni")
plt.ylabel("liczba chomik√≥w")
plt.show()

# --- opcjonalnie: szczyt kariery w podziale na kategorie ---
peak_by_cat = (
    comp.groupby(["category", "hamster_id", "name_hamster"], as_index=False)
        .agg(
            wins=("competition_id", "count"),
            first_win=("competition_date", "min"),
            last_win=("competition_date", "max"),
        )
)
peak_by_cat = peak_by_cat[peak_by_cat["wins"] >= 2].copy()
peak_by_cat["peak_days"] = (peak_by_cat["last_win"] - peak_by_cat["first_win"]).dt.days

summary_cat = peak_by_cat.groupby("category", as_index=False).agg(
    hamsters=("hamster_id", "nunique"),
    mean_peak_days=("peak_days", "mean"),
    median_peak_days=("peak_days", "median")
).sort_values("mean_peak_days", ascending=False)

summary_cat
```

## Wp≈Çyw majƒÖtku w≈Ça≈õciciela na wyniki

```python
# =========================================================
# Przewaga bogatych w≈Ça≈õcicieli: analiza per KATEGORIA i per DYSCYPLINA
# =========================================================

from scipy import stats
import numpy as np

def wealth_advantage_test(df, group_col):
    """
    df        - DataFrame (comp)
    group_col - 'category' lub 'discipline_id'
    Zwraca tabelƒô z wynikami testu Manna‚ÄìWhitneya
    """
    results = []

    for g, sub in df.groupby(group_col):
        # kwartyle majƒÖtku
        q1 = sub["owner_networth"].quantile(0.25)
        q3 = sub["owner_networth"].quantile(0.75)

        bottom = sub[sub["owner_networth"] <= q1]["score"]
        top = sub[sub["owner_networth"] >= q3]["score"]

        if len(top) < 8 or len(bottom) < 8:
            results.append({
                group_col: g,
                "n_top": len(top),
                "n_bottom": len(bottom),
                "top_mean": np.nan,
                "bottom_mean": np.nan,
                "p_value": np.nan,
                "advantage": "za ma≈Ço danych"
            })
            continue

        stat, p = stats.mannwhitneyu(top, bottom, alternative="two-sided")

        advantage = (
            "TAK (bogaci)" if (top.mean() > bottom.mean() and p < 0.05)
            else "TAK (biedni)" if (top.mean() < bottom.mean() and p < 0.05)
            else "NIE"
        )

        results.append({
            group_col: g,
            "n_top": len(top),
            "n_bottom": len(bottom),
            "top_mean": round(top.mean(), 3),
            "bottom_mean": round(bottom.mean(), 3),
            "p_value": round(p, 4),
            "advantage": advantage
        })

    return (
        pd.DataFrame(results)
        .sort_values("p_value", na_position="last")
        .reset_index(drop=True)
    )

# ---------------------------------------------------------
# 1) Analiza per KATEGORIA
# ---------------------------------------------------------

adv_by_category = wealth_advantage_test(comp, "category")
adv_by_category
```

## Popularno≈õƒá sportu (trend)

```python
# =========================================================
# Popularno≈õƒá sportu w czasie
# =========================================================

popularity = (
    comp.groupby("year", as_index=False)
        .agg(
            competitions=("competition_id", "count"),
            participants=("hamster_id", "nunique"),
            avg_audience=("spectators_stadium", "mean"),
            total_audience=("spectators_stadium", "sum"),
        )
        .sort_values("year")
)

popularity

plt.figure()
plt.plot(popularity["year"], popularity["participants"], marker="o")
plt.title("Liczba uczestnik√≥w zawod√≥w (unikalne chomiki)")
plt.xlabel("Rok")
plt.ylabel("Liczba chomik√≥w")
plt.show()

plt.figure()
plt.plot(popularity["year"], popularity["competitions"], marker="o")
plt.title("Liczba zawod√≥w w roku")
plt.xlabel("Rok")
plt.ylabel("Liczba zawod√≥w")
plt.show()

plt.figure()
plt.plot(popularity["year"], popularity["avg_audience"], marker="o")
plt.title("≈örednia widownia stadionowa na zawody")
plt.xlabel("Rok")
plt.ylabel("≈örednia liczba widz√≥w")
plt.show()

```

```python
# =========================================================
# Atrakcyjno≈õƒá kategorii i dyscyplin
# =========================================================

attr_category = (
    comp.groupby("category", as_index=False)
        .agg(
            events=("competition_id", "count"),
            avg_audience=("spectators_stadium", "mean"),
            total_audience=("spectators_stadium", "sum"),
        )
        .sort_values("avg_audience", ascending=False)
)

attr_discipline = (
    comp.groupby(["discipline_id", "name", "category"], as_index=False)
        .agg(
            events=("competition_id", "count"),
            avg_audience=("spectators_stadium", "mean"),
            total_audience=("spectators_stadium", "sum"),
        )
        .sort_values("avg_audience", ascending=False)
)

attr_category, attr_discipline

plt.figure()
plt.bar(attr_category["category"], attr_category["avg_audience"])
plt.title("≈örednia widownia stadionowa wg kategorii")
plt.xlabel("Kategoria")
plt.ylabel("≈örednia liczba widz√≥w")
plt.show()

```

```python
# =========================================================
# Rentowno≈õƒá federacji: przychody vs wydatki w czasie
# =========================================================

import pandas as pd
import matplotlib.pyplot as plt
# Wczytanie danych finansowych z bazy
funding_sources = pd.read_sql("SELECT * FROM funding_sources", conn)
sponsorships = pd.read_sql("SELECT * FROM sponsorships", conn)
employees = pd.read_sql("SELECT * FROM employees", conn)

# konwersja dat
for df in [funding_sources, sponsorships, employees]:
    for c in df.columns:
        if "date" in c:
            df[c] = pd.to_datetime(df[c])

# zakres lat analizy
year_min = comp["year"].min()
year_max = comp["year"].max()

# -------- PRZYCHODY --------

def expand_amounts_by_year(df, start_col, end_col, amount_col):
    rows = []
    for _, r in df.iterrows():
        start = r[start_col].year
        end = r[end_col].year if pd.notna(r[end_col]) else year_max
        for y in range(start, end + 1):
            if y >= year_min and y <= year_max:
                rows.append((y, float(r[amount_col])))
    return pd.DataFrame(rows, columns=["year", "amount"])

# dotacje
funding_y = (
    expand_amounts_by_year(funding_sources, "start_date", "end_date", "amount")
    .groupby("year", as_index=False)["amount"]
    .sum()
    .rename(columns={"amount": "funding"})
)

# sponsoring
sponsorship_y = (
    expand_amounts_by_year(sponsorships, "start_date", "end_date", "contribution_amount")
    .groupby("year", as_index=False)["amount"]
    .sum()
    .rename(columns={"amount": "sponsorship"})
)

# -------- WYDATKI (wynagrodzenia) --------

def payroll_by_year(employees, year_min, year_max):
    rows = []
    for y in range(year_min, year_max + 1):
        y_start = pd.Timestamp(y, 1, 1)
        y_end = pd.Timestamp(y, 12, 31)
        active = employees[
            (employees["start_date"] <= y_end) &
            ((employees["end_date"].isna()) | (employees["end_date"] >= y_start))
        ]
        rows.append((y, active["salary_amount"].sum() * 12))
    return pd.DataFrame(rows, columns=["year", "payroll"])

payroll_y = payroll_by_year(employees, year_min, year_max)

# -------- ZESTAWIENIE --------

finance = (
    pd.DataFrame({"year": range(year_min, year_max + 1)})
    .merge(funding_y, on="year", how="left")
    .merge(sponsorship_y, on="year", how="left")
    .merge(payroll_y, on="year", how="left")
    .fillna(0)
)

finance["income"] = finance["funding"] + finance["sponsorship"]
finance["expenses"] = finance["payroll"]
finance["balance"] = finance["income"] - finance["expenses"]

finance
plt.figure()
plt.plot(finance["year"], finance["income"], marker="o", label="Przychody")
plt.plot(finance["year"], finance["expenses"], marker="o", label="Wydatki")
plt.legend()
plt.title("Przychody i wydatki federacji w czasie")
plt.xlabel("Rok")
plt.ylabel("PLN")
plt.show()
```

```python
# =========================================================
# P1: Wp≈Çyw wieku chomika na wyniki
# =========================================================

comp_age = comp.copy()
comp_age["age_days"] = (comp_age["competition_date"] - comp_age["birth_date"]).dt.days
comp_age = comp_age[comp_age["age_days"] > 0]

from scipy import stats

rho, p = stats.spearmanr(comp_age["age_days"], comp_age["score"])

print(f"Spearman(age, score): rho={rho:.3f}, p={p:.4f}")

import matplotlib.pyplot as plt
plt.figure()
plt.scatter(comp_age["age_days"], comp_age["score"], alpha=0.5)
plt.xlabel("Wiek chomika (dni)")
plt.ylabel("Wynik (score)")
plt.title("Wp≈Çyw wieku chomika na wyniki")
plt.show()
```

```python
# =========================================================
# P2: Czy doping dawa≈Ç przewagƒô przed wykryciem?
# =========================================================

first_positive = (
    doping_controls[doping_controls["is_positive"] == 1]
    .groupby("hamster_id", as_index=False)["control_date"]
    .min()
    .rename(columns={"control_date": "first_positive"})
)

comp_d = comp.merge(first_positive, on="hamster_id", how="left")

before_doping = comp_d[
    (comp_d["first_positive"].notna()) &
    (comp_d["competition_date"] < comp_d["first_positive"])
]["score"]

clean = comp_d[comp_d["first_positive"].isna()]["score"]

from scipy import stats
if len(before_doping) >= 5 and len(clean) >= 10:
    u, p = stats.mannwhitneyu(before_doping, clean)
    print("Mann‚ÄìWhitney p =", p)
    print("≈öredni score (doping przed):", before_doping.mean())
    print("≈öredni score (czyste):", clean.mean())
else:
    print("Za ma≈Ço danych do testu")
```

```python
# =========================================================
# P3 (POPRAWIONE): Czy liczba POTENCJALNYCH start√≥w wp≈Çywa na wyniki?
# =========================================================

from scipy import stats
import pandas as pd
import matplotlib.pyplot as plt

# zakres dat zawod√≥w (globalny)
min_comp_date = comp["competition_date"].min()
max_comp_date = comp["competition_date"].max()

# 1) Wyznaczenie okresu kariery ka≈ºdego chomika
ham_career = hamsters.copy()
ham_career["career_start"] = ham_career["birth_date"] + pd.Timedelta(days=60)
ham_career["career_end"] = ham_career["retire_date"].fillna(max_comp_date)

# zabezpieczenie: nie wykraczamy poza dane
ham_career["career_start"] = ham_career["career_start"].clip(lower=min_comp_date)
ham_career["career_end"] = ham_career["career_end"].clip(upper=max_comp_date)

# 2) Liczba POTENCJALNYCH start√≥w = liczba zawod√≥w w okresie kariery
def count_possible_starts(row):
    return comp[
        (comp["competition_date"] >= row["career_start"]) &
        (comp["competition_date"] <= row["career_end"])
    ]["competition_id"].nunique()

ham_career["possible_starts"] = ham_career.apply(count_possible_starts, axis=1)

# 3) Wyniki sportowe chomika
perf = (
    comp.groupby(["hamster_id", "name_hamster"], as_index=False)
        .agg(
            avg_score=("score", "mean"),
            wins=("competition_id", "count")
        )
)

# 4) Po≈ÇƒÖczenie: potencjalne starty + wyniki
analysis_p3 = ham_career.merge(
    perf, on="hamster_id", how="left"
).dropna(subset=["avg_score"])

analysis_p3[["name_hamster", "possible_starts", "avg_score"]].head()
# 5) Korelacja: potencjalne starty ‚Üî wynik
rho, p = stats.spearmanr(
    analysis_p3["possible_starts"],
    analysis_p3["avg_score"]
)

print(f"Spearman(possible_starts, avg_score): rho={rho:.3f}, p={p:.4f}")
# 6) Wizualizacja
plt.figure()
plt.scatter(
    analysis_p3["possible_starts"],
    analysis_p3["avg_score"],
    alpha=0.6
)
plt.xlabel("Liczba potencjalnych start√≥w (zawody w okresie kariery)")
plt.ylabel("≈öredni wynik (score)")
plt.title("Potencjalne starty a wyniki chomik√≥w")
plt.show()
```

```python
# =========================================================
# P4 (POPRAWIONE): Konkurencyjno≈õƒá dyscyplin
# Znormalizowane odchylenie: std / mean (CV)
# =========================================================

import matplotlib.pyplot as plt

competitiveness = (
    comp.groupby(["discipline_id", "name"], as_index=False)
        .agg(
            events=("competition_id", "count"),
            mean_time=("winner_result", "mean"),
            std_time=("winner_result", "std")
        )
)

# wsp√≥≈Çczynnik zmienno≈õci (CV)
competitiveness["cv"] = competitiveness["std_time"] / competitiveness["mean_time"]

competitiveness = competitiveness.sort_values("cv")

competitiveness
# Wizualizacja konkurencyjno≈õci
plt.figure()
plt.barh(competitiveness["name"], competitiveness["cv"])
plt.xlabel("Wsp√≥≈Çczynnik zmienno≈õci (std / mean)")
plt.title("Konkurencyjno≈õƒá dyscyplin (ni≈ºszy = bardziej wyr√≥wnana)")
plt.show()
```
